```python
# ============================================================
# BLOCK 2: Load both datasets
# ============================================================
import pandas as pd
import numpy as np

# Load Dataset 1 (Students Performance in Exams)
df1 = pd.read_csv('/content/drive/MyDrive/IS_Project/StudentsPerformance.csv')

# Load Dataset 2 (Student Performance Factors)
df2 = pd.read_csv('/content/drive/MyDrive/IS_Project/StudentPerformanceFactors.csv')

print("Dataset 1 shape:", df1.shape)
print("Dataset 2 shape:", df2.shape)
print("\nDataset 1 columns:", df1.columns.tolist())
print("Dataset 2 columns:", df2.columns.tolist())
```

```python
# ============================================================
# BLOCK 3: Explore missing values
# ============================================================
print("=== Dataset 1 Missing Values ===")
print(df1.isnull().sum())

print("\n=== Dataset 2 Missing Values ===")
print(df2.isnull().sum())
# You will see missing values in Teacher_Quality,
# Parental_Education_Level, Distance_from_Home
```

```python
# ============================================================
# BLOCK 4: Fix Dataset 1
# ============================================================

# Rename columns to be easier to work with
df1.columns = [
    'gender', 'race_ethnicity', 'parental_education',
    'lunch', 'test_prep', 'math_score',
    'reading_score', 'writing_score'
]

# Create an average score column
df1['avg_score'] = (df1['math_score'] + df1['reading_score'] + df1['writing_score']) / 3

# Create a Pass/Fail column (pass = avg score >= 60)
df1['result'] = (df1['avg_score'] >= 60).astype(int)
# 1 = Pass, 0 = Fail

print(df1.head())
print("\nPass/Fail counts:\n", df1['result'].value_counts())
```

```python
# ============================================================
# BLOCK 5: Fix Dataset 2 — Handle Missing Values
# ============================================================

# Check how many missing values exist
print("Missing before fix:")
print(df2[['Teacher_Quality','Parental_Education_Level','Distance_from_Home']].isnull().sum())

# Fill text/category columns with the most common value (mode)
df2['Teacher_Quality'] = df2['Teacher_Quality'].fillna(df2['Teacher_Quality'].mode()[0])
df2['Parental_Education_Level'] = df2['Parental_Education_Level'].fillna(df2['Parental_Education_Level'].mode()[0])
df2['Distance_from_Home'] = df2['Distance_from_Home'].fillna(df2['Distance_from_Home'].mode()[0])

print("\nMissing after fix:")
print(df2[['Teacher_Quality','Parental_Education_Level','Distance_from_Home']].isnull().sum())
```

```python
# ============================================================
# BLOCK 6: Encode text columns to numbers (Dataset 2)
# ============================================================
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# These columns contain text like "High", "Low", "Yes", "No"
text_cols = [
    'Parental_Involvement', 'Access_to_Resources', 'Extracurricular_Activities',
    'Motivation_Level', 'Internet_Access', 'Family_Income',
    'Teacher_Quality', 'School_Type', 'Peer_Influence',
    'Learning_Disabilities', 'Parental_Education_Level',
    'Distance_from_Home', 'Gender'
]

for col in text_cols:
    df2[col] = le.fit_transform(df2[col])

print("Encoding done!")
print(df2.head())
```

```python
# ============================================================
# BLOCK 7: Scale numeric columns (Dataset 2)
# ============================================================
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

num_cols = [
    'Hours_Studied', 'Attendance', 'Sleep_Hours',
    'Previous_Scores', 'Tutoring_Sessions', 'Physical_Activity'
]

df2[num_cols] = scaler.fit_transform(df2[num_cols])

print("Scaling done!")
```

```python
# ============================================================
# BLOCK 8: Create Pass/Fail column for Dataset 2
# ============================================================

# Exam_Score >= 60 = Pass
df2['result'] = (df2['Exam_Score'] >= 60).astype(int)

print("Pass/Fail counts:\n", df2['result'].value_counts())
```

```python
# ============================================================
# BLOCK 9: Save cleaned datasets
# ============================================================
import pickle

df1.to_csv('/content/drive/MyDrive/IS_Project/cleaned_df1.csv', index=False)
df2.to_csv('/content/drive/MyDrive/IS_Project/cleaned_df2.csv', index=False)

# Save the scaler (needed later in the web app)
with open('/content/drive/MyDrive/IS_Project/scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

print("✅ All saved!")
```

```python
# ============================================================
# ADD THIS AS A NEW BLOCK at the end of data_preparation.ipynb
# Run this AFTER Block 6 (the encoding block)
# This saves the label encoders so the website can use them
# ============================================================

import pickle

# Save all label encoders into one dictionary
# We need to re-fit them here to save them properly

from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Reload the original (before encoding) to re-fit encoders
df2_original = pd.read_csv('/content/drive/MyDrive/IS_Project/StudentPerformanceFactors.csv')

# Fill missing values first (same as before)
df2_original['Teacher_Quality'] = df2_original['Teacher_Quality'].fillna(df2_original['Teacher_Quality'].mode()[0])
df2_original['Parental_Education_Level'] = df2_original['Parental_Education_Level'].fillna(df2_original['Parental_Education_Level'].mode()[0])
df2_original['Distance_from_Home'] = df2_original['Distance_from_Home'].fillna(df2_original['Distance_from_Home'].mode()[0])

text_cols = [
    'Parental_Involvement', 'Access_to_Resources', 'Extracurricular_Activities',
    'Motivation_Level', 'Internet_Access', 'Family_Income',
    'Teacher_Quality', 'School_Type', 'Peer_Influence',
    'Learning_Disabilities', 'Parental_Education_Level',
    'Distance_from_Home', 'Gender'
]

label_encoders = {}
for col in text_cols:
    le = LabelEncoder()
    le.fit(df2_original[col])  # fit on original text values
    label_encoders[col] = le
    print(f"{col}: {list(le.classes_)}")  # shows what each column's options are

# Save all encoders
with open('/content/drive/MyDrive/IS_Project/label_encoders.pkl', 'wb') as f:
    pickle.dump(label_encoders, f)

print("\n✅ Label encoders saved!")
print("Download label_encoders.pkl from Google Drive to your IS_Project folder")
```
